```python
import pandas as pd
import os

bids_csv = pd.read_csv("/home/jaimebarranco/Desktop/MRI-QC/fetal/fetalqc_non-labeled-dataset/bids_csv.csv")

# get the subset of bids_csv where 'name' is present
bids_csv = bids_csv[bids_csv['name'].notnull()]

# subdataframe from bids_csv with 'name' and 'sub' columns
df_aux = bids_csv[['name', 'sub']]

# order by 'sub'
df_aux = df_aux.sort_values(by=['sub'])

subjects_names = sorted(os.listdir('/mnt/sda1/Repos/a-eye/Data/SHIP_dataset/non_labeled_dataset_nifti'))

# add subjects_names to df_aux
df_aux['subjects_names'] = subjects_names

# change column name 'name' for 'report', 'sub' for 'bids' and 'subjects_names' for 'subject'
df_aux = df_aux.rename(columns={'name': 'report', 'sub': 'bids', 'subjects_names': 'subject'})

# add 'sub-' to 'bids' column having 3 numbers in total for each case
df_aux['bids'] = df_aux['bids'].apply(lambda x: f'sub-{x:03}')

# save df_aux to csv
df_aux.to_csv('/home/jaimebarranco/Desktop/MRI-QC/df_aux1.csv', index=False)
```
