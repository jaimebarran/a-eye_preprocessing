```python
''' Copy reports from one folder to another based on an excel file'''

import pandas as pd
import os

excel_file = pd.read_excel('/mnt/sda1/Repos/a-eye/Output/mri_qc/excluded_edited.xlsx')
input_path = '/mnt/sda1/Repos/a-eye/Output/mri_qc/fetal/fetalqc_non-labeled-dataset'
output_path = '/mnt/sda1/Repos/a-eye/Output/mri_qc/fetal/fetal_qc_100'

subject_names = excel_file['subject_name'].tolist()
subject_names = [sub + '_report.html' for sub in subject_names]  # add _report.html to the subject names

for sub in subject_names:
    src = os.path.join(input_path, sub)
    dst = os.path.join(output_path, sub)
    if os.path.exists(src):
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        os.system(f'cp -r {src} {dst}')
    else:
        print(f"Source folder {src} does not exist.")
```
